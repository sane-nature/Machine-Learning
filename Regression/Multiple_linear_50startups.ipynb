```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df= pd.read_csv('data/50_Startups.csv')
```

```python
#Dividing dependent and independent variables.
x= df.drop('Profit', axis=1)
y= df['Profit']
x
```

```python
dummies= pd.get_dummies(x.State)
dummies
```

```python
merged= pd.concat([x, dummies], axis= 'columns')
merged.drop(['State','California'], axis='columns', inplace=True)
merged
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test= train_test_split(merged,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
clf= LinearRegression()
clf.fit(x_train,y_train);
```

```python
#Predicting the results
y_preds= clf.predict(x_test)
y_preds
```

```python
#Backward elimination
import statsmodels.api as sm
merged=np.append(arr=np.ones((50,1)).astype(int), values= merged , axis=1)
```

```python
x_opt=merged[:,[0,1,2,3,4,5]]
regressor_OLS=sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt=merged[:,[0,1,2,3,4]]
regressor_OLS=sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt=merged[:,[0,1,2,4]]
regressor_OLS=sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt=merged[:,[0,1,2]]
regressor_OLS=sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```
